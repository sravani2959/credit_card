```python
def whisker(col):
    q1, q3 = np.percentile(col, [25, 75])
    iqr = q3 - q1
    lw = q1 - 1.5 * iqr
    uw = q3 + 1.5 * iqr
    return lw, uw
```

```python
df2.columns
```

```python
whisker(df2['NPA Status'])
```

```python
whisker(df2['RevolvingUtilizationOfUnsecuredLines'])
```

```python
for i in ['NPA Status']:
       lw, uw = whisker(df2[i])
       df2[i] = np.where(df2[i] < lw, lw, df2[i])
       df2[i] = np.where(df2[i] > uw, uw, df2[i])
```
